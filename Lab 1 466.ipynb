```python

india_grouped = india.groupby('Year')['Count'].sum().reset_index()
italy_grouped = italy.groupby('Year')['Count'].sum().reset_index()

merged_df = pd.merge(india_grouped, italy_grouped, on='Year', suffixes=('_india', '_italy'))

plt.figure(figsize=(10, 6))
plt.plot(merged_df['Year'], merged_df['Count_india'], label='India')
plt.plot(merged_df['Year'], merged_df['Count_italy'], label='Italy')


merged_df['Total_Count'] = merged_df['Count_india'] + merged_df['Count_italy']
plt.plot(merged_df['Year'], merged_df['Total_Count'], label='Total Count', linestyle='--')

plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Counts per Year for India and Italy')
plt.legend()

plt.show()
```
