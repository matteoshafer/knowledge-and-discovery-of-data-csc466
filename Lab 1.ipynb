## Research Question 3

What are the most common baby names in each state and how do they differ from national trends? How do the most popular names length change over time?

```python
state["length"] = state.Name.apply(len)
N = 10

popularOnly = state.groupby(["State", "Year"])["Count"].apply(lambda g: g.sort_values(ascending=False).head(N)).reset_index()
popularOnly = popularOnly.merge(state, on=["State", "Year", "Count"]).drop_duplicates(subset=["State", "Year", "Count", "Name"])
popularOnly
```

As a quantifier for the trends, we will compare the lengths over time.

```python
f'The top {N} baby names in each state are in the above table per year. This does not take gender into account'
```

```python
grouped = popularOnly.groupby("Year")["length"].mean()
grouped = grouped.reset_index()
fig = px.line(grouped, x="Year", y="length", title = 'How has the mean baby name length changed over time?', labels={'Year': 'Year', 'length': 'mean length'})
fig.show()
```

```python
# fit a linear regression model to estimate the average change in length over year
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import *
```

```python
X = grouped.drop("length", axis=1)
y = grouped["length"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)
model.intercept_, model.coef_[0]
```

```python
r2_score(y_test, model.predict(X_test))
```

### State by state summary?

```python
import plotly.subplots as subplots
import plotly.graph_objs as go
from plotly.offline import iplot, init_notebook_mode
```

```python
grouped = popularOnly.groupby(["State", "Year"])["length"].mean()
grouped
```

```python
grouped = grouped.reset_index()
```

```python
# make a seaborn facet grid for each state and the mean length over time
import seaborn as sns
sns.FacetGrid(grouped, col="State", col_wrap=5, height=2).map(sns.lineplot, "Year", "length")
```

```python
def linearRegressionStates(stateName):
    filtered = grouped[grouped["State"] == stateName]
    X = filtered[["Year"]]
    y = filtered["length"]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
    model = LinearRegression()
    model.fit(X_train, y_train)
    return model.coef_[0]
```

```python
# map the above function over the unique states
slopes = pd.Series(state["State"].unique()).apply(linearRegressionStates)
slopes.plot.density(xlabel="Average Yearly Mean Length Change")
```
